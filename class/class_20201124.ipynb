```python
df = pd.read_csv("../data/diamonds.csv")
print(df.head())
print()
price = df["price"]
# df["price"] = (price - min(price)) / (max(price) - min(price))
df["price"] = (price - price.min()) / (price.max() - price.min())

print()
print(df["price"].max())
print()
print(df["price"].min())
print()
```

```python
depth = df["depth"]
df["depth"] = (depth - depth.mean()) / depth.std()

print(df["depth"].max())
print(df["depth"].min())
```

```python
def normalization(series=None):
    if series is not None:
        return (series - series.min()) / (series.max() - series.min())

def standardization(series=None):
    if series is not None:
        return (series - series.mean()) / series.std()
    
print(normalization(df["depth"]).head())
print()
print(standardization(df["depth"]).head())
print()

print(normalization(pd.Series([2,3,5,7,10])))
print()
print(standardization(pd.Series([2,3,5,7,10])))
print()
```
