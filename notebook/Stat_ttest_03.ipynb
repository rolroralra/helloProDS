### Q1. Pandas, Scipy의 stats 모듈을 불러오시오.

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
```

### Q2. "bike.csv" 파일을 불러와서 df 객체에 저장하시오.

```python
df = pd.read_csv("../data/bike.csv")
df.info()
df.head()
```

### Q3. df객체의 "datetime"변수에서 월 정보를 뽑아 df객체에 "month" 변수를 새로 만드시오.

```python
df["month"] = pd.to_datetime(df["datetime"]).dt.month
df
```

### Q4. 6월과 12월의 자전거 대여 차이를 알아보고자 한다. "registered" 변수와 독립 2표본 t검정을 사용하여 결과를 확인하시오.

```python
df_6_month = df.loc[df["month"] == 6, :]
df_12_month = df.loc[df["month"] == 12, :]

stats.ttest_ind(df_6_month["registered"],df_12_month["registered"])
```

### Q5. df객체의 "datetime"변수에서 요일 정보를 뽑아 df객체에 "wday" 변수를 새로 만드시오

```python
df["wday"] = pd.to_datetime(df["datetime"]).dt.weekday
# df["wday"] = pd.to_datetime(df["datetime"]).dt.dayofweek
df
```

### Q6. 주말(토, 일)이 1이고 나머지 요일이 0인 "wend" 변수를 df객체에 만드시오.

```python
# df["wend"] = df["wday"].apply(lambda x: 1 if x > 4 else 0)
df["wend"] = (df["wday"] > 4) + 0
df
```

### Q7. 주중과 주말에 따른 "casual" 변수값의 차이가 있는지 독립 2표본 t검정을 사용하여 결과를 확인하시오.

```python
df_weekday = df.loc[df["wend"] == 0, "casual"]
df_weekend = df.loc[df["wend"] == 1, "casual"]
stats.ttest_ind(df_weekday, df_weekend)
```
