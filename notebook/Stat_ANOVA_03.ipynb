### Q1. Pandas, statsmodels 모듈을 불러오시오.

```python
import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

### Q2. "diabetes.csv" 파일을 읽어와서 df 객체에 저장하시오.

```python
df = pd.read_csv("../data/diabetes.csv")
df.head(5)
```

### Q3. 혈압이 30 이하인 측정 데이터는 모두 제거하여 df_up30 에 저장하시오

```python
df_up30 = df.loc[df["BloodPressure"] > 30, :].copy().reset_index(drop=True)
df_up30

# df_up30_2 = df.drop(df[df["BloodPressure"] <= 30].index, axis=0).copy().reset_index(drop=True)
# df_up30_2
```

### Q4. 나이를 10살 단위로 구분한 "Age_group" 변수를 만드시오
※ 예시: 7 => 0, 30 => 3, 45 => 4

```python
# df_up30["Age_group"] = df_up30["Age"].apply(lambda x: x // 10)
df_up30["Age_group"] = df_up30["Age"] // 10
df_up30
```

### Q5. 연령대별 혈당(Glucose)의 차이가 있는지 일원분산분석을 실시하고 그 결과를 확인하시오.

```python
# lm = ols("Glucose ~ Age_group", df).fit()
lm = ols("Glucose ~ C(Age_group)", df_up30).fit()
anova_lm(lm)
```

### Q6. 일원분산분석의 사후검정을 실시하시오. 단, Tukey's HSD 를 실시

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

posthoc = pairwise_tukeyhsd(df_up30["Glucose"], df_up30["Age_group"])
print(posthoc)
```
