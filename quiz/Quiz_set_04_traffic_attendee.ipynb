### Q9. 출발지점(StartPoint)과 도착지점(column)의 일 교통량의 최대값을 "max" 라는 이름의 변수에 저장하시오.


```python
# df["max"] = df.iloc[:,2:].max(axis=1)   # 조심... start:end 정해져야됨.
df["max"] = df.loc[:, "Gyeonggi":"Gangwon"].max(axis=1)
df
```

### Q10. 일별 출발지점별 총 통행량을 "sum" 이라는 이름의 변수에 저장하시오.


```python
# df["sum"] = df.iloc[:,2:].sum(axis=1) # Error
df["sum"] = df.loc[:, "Gyeonggi":"Gangwon"].sum(axis=1)
df
```

### Q11. [max] 변수를 참고하여 일별 출발지점별 교통량이 가장 많았던 일자와 그 출발지점을 기술하시오.


```python
df.iloc[df["max"].idxmax(),0:2]
# df.iloc[df["max"].idxmax(),[0,1]]
df.loc[[df["max"].idxmax()],["date","StartPoint"]]
```

### Q12. [sum] 변수를 참고하여 일별 출발지점별 교통량이 가장 많았던 일자와 그 출발지점을 기술하시오.


```python
df.iloc[[df["sum"].idxmax()],[0, 1]]
```

### Q13. 출발지점별 일 최대 교통량을 구하시오.


```python
df["date"] = pd.to_datetime(df["date"], format="%Y%m%d")
df["month"] = df["date"].dt.year * 100 + df["date"].dt.month
df.groupby(by=["month", "StartPoint"]).agg(["max"])["max"].reindex()
# df.groupby(by=["date","StartPoint"])
df
```
