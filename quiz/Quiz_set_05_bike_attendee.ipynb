### Q7. pandas.DataFrame의 join() 을 활용하여 weather과 rental 데이터를 병합하시오.
(병합한 데이터는 df_join에 저장하시오)
 * pandas.to_timedelta() 를 활용하시오
 * pandas.Series.astype() 를 활용하시오

```python
rental = pd.read_csv("../data/bike_rental.csv")
weather = pd.read_csv("../data/bike_weather.csv")

# rental["datetime"] = pd.to_datetime(rental["datetime"])


# rental["date"] = rental["datetime"].dt.date
# rental["hour"] = rental["datetime"].dt.hour

# rental = rental.set_index(["date", "hour"])
# # rental

# # weather = weather.set_index(["date","hour"])
# weather.index
# # df_join = weather.join(rental, on=["date", "hour"])
# # df_join
# # df_join.head()


# Way2. Using by pandas.to_timedelta(...)
rental["datetime"] = pd.to_datetime(rental["datetime"])
weather["datetime"] = pd.to_datetime(weather["date"] + " " + weather["hour"].astype("string").str.pad(2, fillchar="0") + ":00:00")

rental["timedelta"] = pd.to_timedelta(rental["datetime"] - rental["datetime"].min())
weather["timedelta"] = pd.to_timedelta(weather["datetime"] - weather["datetime"].min())

weather = weather.drop(["datetime"], axis=1)
df_join = weather.set_index("timedelta").join(rental.set_index("timedelta"))
df_join

# weather["hour"].astype("string").str.pad(2, fillchar="0", side="left")
# weather["hour"].astype("string").str.zfill(2)


# rental = rental.drop(["timedelta"], axis=1)
# weather = weather.drop(["timedelta", "datetime"], axis=1)

# df_join = weather.join(rental, lsuffix="w", rsuffix="r")
# df_join
```

### Q9. 정기권 이용자의 자전거 대여[registered] 총합이 가장 많은 계절[season]은 언제인가?
(1: 봄, 2: 여름, 3: 가을, 4: 겨울)
 * df_join 객체 기준으로 풀이하시오

```python
df_join.groupby(by=["season"]).agg(["sum"])["registered"].idxmax()
```
